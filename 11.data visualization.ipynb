```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("./data/시도별 전출입 인구수.xlsx")
```

## 결측치 처리

```python
df.dropna()
```

```python
df = df.fillna(method='ffill')
```

```python
df
```

## 데이터 추출

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
```

```python
df[mask]
```

```python
df_seoul = df[mask]
```

```python
# 전출지별 컬럼 삭제
df_seoul = df_seoul.drop(['전출지별'], axis =1)
```

```python
df_seoul.head()
```

```python
# 전입지별 -> 전입지, 컬럼명 변경
df_seoul = df_seoul.rename(columns = {'전입지별' : '전입지'})
```

```python
df_seoul.head()
```

```python
# 인덱스 '전입지'로 변경
df_seoul = df_seoul.set_index('전입지')
```

```python
sr_one = df_seoul.loc['경기도']
```

## 시각화

```python
plt.plot(sr_one.index, sr_one.values)
plt.show()
```

## 그림 디자인 설정

```python
print(plt.style.available)
```

```python
# 스타일
plt.style.use('ggplot')

# 그림 크기
plt.figure(figsize=(14,5))

# x축의 각도
# plt.xticks(rotation='vertical')
plt.xticks(rotation='45')

# 그림 제목
plt.title("서울 → 경기 인구 이동",fontdict={'fontsize' : 15})

# 축 이름
plt.xlabel("기간")
plt.ylabel("인구 이동 수")

# 축 범위 지정, ylim(), xlim()
plt.ylim(50000, 800000)

# 그림 그리기
plt.plot(sr_one) #시리즈일 경우 그래프로 자동 생성됨

# legend
plt.legend(labels=['서울 → 경기'], fontsize = 15)

# 주석 표시 - 화살표
plt.annotate('',
             xy=(20, 620000),       #화살표의 머리 부분(끝점)
             xytext=(2, 290000),    #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='skyblue', lw=5), #화살표 서식
             )

plt.annotate('',
             xy=(47, 450000),       #화살표의 머리 부분(끝점)
             xytext=(30, 580000),   #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='olive', lw=5),  #화살표 서식
             )

plt.annotate("인구이동 증가", 
             xy = (10, 470000),
             rotation=22,
             va = 'baseline', # 텍스트 상하 정렬
             ha='center',  # 텍스트 좌우 정렬
             fontsize=15)

plt.annotate("인구이동 감소", 
             xy = (39, 550000),
             rotation = -8,
             va = 'baseline', # 텍스트 상하 정렬
             ha='center',  # 텍스트 좌우 정렬
             fontsize=15)

plt.show()
```

## 시각화2

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# 그래프 객체 생성
fig = plt.figure(figsize=(14,14))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

# 그림 제목
ax1.set_title("그림 1", size = 15)
ax2.set_title("그림 2", size = 15)

# 그림 마커 설정
ax1.plot(sr_one, 'o', markersize=10)
ax2.plot(sr_one, marker='*', markerfacecolor='green', markersize=10, 
         color='olive', linewidth = 5, label = '서울 → 경기')

# 그림 범주
ax2.legend(loc='best')

# 축 범위 설정
ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000, 800000)

ax1.set_xlabel("서울 → 경기 인구이동")
ax1.set_ylabel("이동 인구 수")
ax2.set_ylabel("이동 인구 수")

ax1.set_xticklabels(sr_one.index, rotation=45)
ax2.set_xticklabels(sr_one.index, rotation=45)

# 축의 라벨 크기 변경
ax1.tick_params(axis='x', labelsize = 8)

plt.show()
```
