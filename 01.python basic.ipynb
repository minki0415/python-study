## 조건문

```python
num = int(input("숫자 값을 입력하세요 : "))

if(num <= 5):
    print("걸어가세요!")
elif(num <= 10):
    print("버스타세요!")
elif(num > 10):
    print("택시타세요!")
else:
    print("잘못된 입력입니다.")
```

# list comprehension

```python
[x for x in range(1,10) if x % 2 == 0]
```

```python
a = []
for x in range(1, 10):
    if x % 2 == 0:
        a.append(x)
    else:
        a.append("!")

a
```

```python
[x if x% 2 == 0 else "1" for x in range(1,10)]
```

```python
[name for name in produce if name[0] != "김"]
```

```python
["{} x {} = {}".format(x, y, x*y) for x in range(2, 10) for y in range(1, 10)]
```

```python
produce[0] + "님"  
```

```python
[name +"님" for name in produce]
```
