# 2019년 서울시 주유소 가격 데이터 

### 1. 각 구마다 가장 많은 브랜드 출력 (내 풀이)

```python
df_list = []
df_list.append(pd.read_csv("../data/상반기 주유소 판매가격.csv", encoding='euc-kr'))
df_list.append(pd.read_csv("../data/하반기 주유소 판매가격.csv", encoding='euc-kr'))

df1 = pd.concat(df_list)
```

```python
df1.shape
```

```python
df1.head(5)
```

```python
df1 = df1[['번호','지역','상표']].drop_duplicates(keep='first', ignore_index=True)
```

```python
df1 = df1.groupby(['지역','상표'], as_index=False).count().rename(columns={'번호':'개수'})
```

```python
df1
```

```python
# 개수별 내림차순 정렬
df1 = df1.sort_values(by=['개수'], ascending=False).reset_index(drop=True)

# 각 구마다 가장 개수가 많은 브랜드 출력
df1.groupby(['지역']).first()
```

# 아파트 실거래가 데이터

## 1. 각 지역별 가장 비싼 아파트

```python
os.listdir("../data/apt")[0]
```

```python
df_apt_list = []
for i in range(12):
    df_apt_list.append(pd.read_csv("../data/apt/"+ os.listdir("../data/apt")[i], encoding="euc-kr", skiprows = 15))
```

```python
df_apt = pd.concat(df_apt_list)
```

```python
df_apt.shape
```

```python
df_apt.head()
```

```python
df_apt1 = df_apt[['시군구', '거래금액(만원)']]
```

```python
df_apt1['시군구'] = df_apt1['시군구'].apply(lambda x : x.split(" ")[0] + " "+ x.split(" ")[1]).copy()
```

```python
df_apt1['거래금액(만원)'] = df_apt1['거래금액(만원)'].apply(lambda x : int(x.replace(",",""))).copy()
```

```python
df_apt1.head()
```

```python
df_apt1.sort_values(by = ['거래금액(만원)'], ascending = False).groupby('시군구').first()
```
