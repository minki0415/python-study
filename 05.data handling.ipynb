#### 선수 각각은 dict, dict는 list에 담는다
- [ {선수1}, {선수2}.... ]

```python
# 내 방법
kbo_player_url = "https://www.koreabaseball.com/Record/Player/HitterDetail/Basic.aspx?playerId={}"
kbo_list = []
for id_ in kbo :
    r = requests.get(kbo_player_url.format(id_))
    bs = BS(r.text)
    li_list = bs.find("div", class_="player_basic").find("ul").findAll("li")
    d = {li.text.split(": ")[0] : li.text.split(": ")[1] for li in li_list}
    kbo_list.append(d)
print(kbo_list)
    
```

```python
len(kbo_list)
```

```python
# 강사님 방법
player_all = []
for idx, id_ in enumerate(kbo) :
    r =requests.get(kbo_player_url.format(id_))
    player_dict = {}
    bs = BS(r.text)
    for x in bs.find("div", class_="player_basic").findAll("li") :
        a, b = x.text.split(":")
        player_dict[a] = b.strip()
    player_all.append(player_dict)
print(player_all)
```

```python
len(player_all)
```

## 지금 선수들 정보에서 선수들의 BMI 값을 추가
- 지금 선수들 정보에서 선수들의 BMI 값을 추가
- 자신의 몸무게(Kg)를 키의 제곱(m)으로 나눈 값

```python
# 키
player_all[0]['신장/체중'].split('/')[0].replace("cm", "")

# 몸무게
player_all[5]['신장/체중'].split('/')[1].replace("kg", "")
```

```python
# 내 방법
for x in player_all:
    x['BMI'] = round(int(x['신장/체중'].split('/')[1].replace("kg", "")) / (int(x['신장/체중'].split('/')[0].replace("cm", ""))/100) ** 2, 3)
```

```python
# 강사님 방법
for x in player_all:
    height, weight = [int(a.replace("cm", "").replace("kg","")) for a in x['신장/체중'].split("/")]
    x['BMI'] = weight / (height / 100) ** 2
```

```python
player_all[1]
```

```python
maximum = 0
rt = dict()

for x in player_all:
    if maximum < x['BMI']:
        maximum = x['BMI']
        rt['maximum'] = [x['선수명'], x['BMI']]
```

```python
from datetime import datetime
```

```python
# 나이 key, value 추가
for x in player_all:
    x['나이'] = datetime.today().year - int(x['생년월일'][:4]) +1
```

```python
# 평균나이
age = 0
for x in player_all:
    age += x['나이']
    
print(age/len(player_all))
```

```python
sum([x['나이'] for x in player_all]) / len([x['나이'] for x in player_all])
```

```python
# 포지션별 등록 인원 수
position = {}
for x in player_all:
    if(x['포지션'] in position):
        position[x['포지션']] += 1
    else:
        position[x['포지션']] = 1

position
```
