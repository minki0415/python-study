```python
import requests
```

```python
keb_url = "https://kebhana.com/cms/rate/wpfxd651_01i_01.do"

payload = {"ajax" : "true",
"curCd" : "",
"tmpInqStrDt" : "2022-04-11",
"pbldDvCd" : "0",
"pbldSqn" : "",
"hid_key_data" : "",
"inqStrDt" : "20220411",
"inqKindCd" : "1",
"hid_enc_data" : "",
"requestTarget" : "searchContentDiv",}

r = requests.post(keb_url, data=payload)
```

```python
money = pd.read_html(r.text)[0]
```

```python
money
```

```python
# 판매할 때 달러환율 값 추출
money.iloc[0,3]
```

```python
all_date = [str(date(2022,1,1) + timedelta(days = x)) for x in range(0, 90)]    
```

```python
for date_ in all_date:
    payload['tmpInqStrDt'] = date_
    payload['inqStrDt'] = date_.replace("-", "")
    r = requests.post(keb_url, data=payload)
    money = pd.read_html(r.text)[0]
    print(money.iloc[0,3])
```
