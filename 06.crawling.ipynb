## 사용자 함수 만들기
- 사용자한테 종목 코드와 페이지 번호를 받으면 데이터를 파싱해서 리스트 형태로 전달

```python
def get_stock_day(code, page):
    url = "https://finance.naver.com/item/sise_day.naver?code={}&page={}".format(code, page)
    head = {"user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36"}
    r = requests.get(url, headers = head)
    bs = BeautifulSoup(r.text, 'lxml')
    
    
    rt = [[x.text.strip() for x in x.findAll("td")] for idx, x in enumerate(bs.find("table").findAll("tr")[2:]) if len(x.findAll("td")) > 1]
    
    return rt
```

```python
get_stock_day("005930", "2")
```

```python
# append 는 x 자체를 원소로 넣고, extend는 iterable의 각 항목을 넣음

total=[]
for x in range(1, 11):
    total.extend(get_stock_day("005930", x))

total
```
